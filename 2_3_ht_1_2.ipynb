```python
#1 Пересечение 

def cross (a, b):
    r = []
    for i in a:
        for k in b:
            if i == k:
                if i not in r:
                    r.append(i)
    return (r)
```

```python
a_b = cross(a,b)
print (a_b)
print (cross (c, a_b))
```

```python
#2 Объединение 

def unite (a, b):
    r = a + b
    return set(r)
```

```python
print (unite(a,b))
```

```python
#3 Разность 

def difference (a,b):
    r = []
    for i in a:
        if i not in b:
            r.append(i)
    return r
```

```python
print (difference(a,b))
```

```python
#4 Симметрическая разность 

def sim_difference (a,b):
    r1 = a + b
    r = [x for n, x in enumerate(r1) if x not in a[:n]]
    return r
```

```python
print (sim_difference(a,b))
```

```python
print (a, b)
```

```python
#5 Декартово произведение 

def dmulitipe (a, b):
    r = []
    for i in a:
        for k in b:
            f = [i, k]
            r.append(f)
    return r
```

```python
print (dmulitipe (a, b))
```

```python
def a1(n):
    a1 = 2**n - n 
    return a1

def b1(n):
    b1 = 1/(1-n)
    return b1

def c1(n):
    c1 = -1**n + np.sqrt(2*n)
    return round (c1,2)

def d1(n):
    d1 = -1**(2*n) + 1/n**2
    return d1
```

```python
# Ответ 3.1 Исследуем монотонность.

for i in range (2,5): 
    print (d1(i))

'''
Резюме: 

Последовательность a_n, b_n - монотонна, а d_n - немонотонна, так как при возрастании, результат предыдущий n больше n+1

'''
```

```python
# Вариант 1. 

def limit(n):
    b = np.math.factorial(n)
    c = pow (b, 1/ n) # корень числа b в n степени 
    return round(c,7)
```

```python
'''
Вариант 1 работает только до n = 100, дальше уже пишет "int too large to convert to float"
'''
print (limit(100)) 
```

```python
'''
Усовершествуем алгоритм и его мощность в вариант 2

'''
# Вариант №2 

import math
from decimal import Decimal

def limit2(n):
    with localcontext() as ctx:
        ctx.prec = 7  # желаемая точность 
        f = Decimal(math.factorial(n))
        c = pow (f, 1/ Decimal(n))
        return c
```

```python
'''
В варианте 2 мы можем расчитывать значение при n > 100, например 2000 
'''
print (limit2(2000))
```
