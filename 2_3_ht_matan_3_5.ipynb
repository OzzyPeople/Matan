**Исследуем интервалы монотонности функции.** Для этого найдем экстремумы функции при помощи производной

**Исследуем функцию на четность и нечетность, периодичность**

Условия **четной** функции - функция, меняющая значение на противоположное при изменении знака независимой переменной (график её симметричен относительно центра координат).

\begin{align}
f(-x) = f(x)
\end{align}

Условие **нечетной** функции - функция, не изменяющая своего значения при изменении знака независимой переменной (график её симметричен относительно оси ординат).

\begin{align}
f(-x) = -f(x)
\end{align}

Проверим нашу функцию

\begin{align}
f(2) = 4 \\
f(-2) = -12 \\
f(-2) \neq - f(2)
\end{align}

Функция не является ни четной, ни нечетной. Очевидно, что функция непериодическая.
